```python
# Criando uma lista com todos os resultados
lista_resultados = []
for dado in dados_api:
  for item in dado['results']:
    dict_dados = {
        'ID_Filme': item['id'],
        'Titulo': item['title'],
        'Titulo Original': item['original_title'],
        'Média do filme': item['vote_average'],
        'ID do genero': item['genre_ids']
    }
    lista_resultados.append(dict_dados)
print(lista_resultados)
```

```python
# Criando lista com resultados
lista_resultados_gen = []

for genero in resultado_gen['genres']:
  dict_gen = {
      'ID': genero['id'],
      'Genero': genero['name']
  }
  lista_resultados_gen.append(dict_gen)
```

```python
# Criando os DataFrames
filmes = pd.DataFrame(lista_resultados)
generos = pd.DataFrame(lista_resultados_gen)
```

```python
display(filmes)
```

```python
filmes = filmes.explode('ID do genero')
filmes = filmes.rename(columns={'ID do genero':'ID'})
```

```python
display(filmes)
```

```python
df_merge = filmes.merge(generos,left_on='ID', right_on='ID', how='left')
```

```python
melhores_filmes = filmes.loc[filmes['Média do filme'] >= 8]
piores_filmes = filmes.loc[(filmes['Média do filme'] <5) & (filmes['Média do filme'] > 0)]
```

```python
# Visualizando graficamente os filmes mais bem avaliados
import matplotlib.pyplot as plt
import numpy as np

# Melhores Filmes:
fig, ax1 = plt.subplots(figsize=(8,8))
ax1.plot(melhores_filmes['Titulo'],melhores_filmes[ 'Média do filme'] )
ax1.set_title('Melhores Filmes')
ax1.set_ylabel('Média dos filmes')
plt.xticks(rotation=90)



plt.show()
```

```python

# Piores filmes
fi2, ax2 = plt.subplots(figsize=(9,8))
ax2.pie(piores_filmes['Média do filme'], labels=piores_filmes['Titulo'], autopct='%1.1f%%', startangle=90)
ax2.legend(piores_filmes['Média do filme'])
ax2.set_title('Piores filmes')
ax2.set_ylabel('Nome dos filmes')
plt.tight_layout()
```
